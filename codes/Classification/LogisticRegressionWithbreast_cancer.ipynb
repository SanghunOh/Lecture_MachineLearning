```python
from sklearn.datasets import load_breast_cancer
import numpy as np
```

```python
cancer = load_breast_cancer()
cancer
```

```python
x = cancer.data
```

```python
y = cancer.target
```

```python
import pandas as pd
pd.DataFrame(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
figure, axes = plt.subplots(ncols=2, nrows=2)
figure.set_size_inches(12, 10)

sns.distplot(x[1], ax=axes[0][0])
sns.distplot(x[2], ax=axes[0][1])
sns.distplot(x[3], ax=axes[1][0])
sns.distplot(x[4], ax=axes[1][1])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = scaler.fit_transform(x)

figure, axes = plt.subplots(ncols=2, nrows=2)
figure.set_size_inches(12, 10)
sns.distplot(x[1], ax=axes[0][0])
sns.distplot(x[2], ax=axes[0][1])
sns.distplot(x[3], ax=axes[1][0])
sns.distplot(x[4], ax=axes[1][1])
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=777, \
                                                    test_size = 0.3, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
log.coef_, log.intercept_
```

```python
y_pred = log.predict(x_test)	# predict í™•ì¸
y_pred
```

```python
predict_proba = log.predict_proba(x_test)	# ì •í™•ë„ í™•ì¸
predict_proba
```

```python
import numpy as np
np.sum(predict_proba, axis=1)
```

```python
y_pred[0], predict_proba[0], predict_proba[0].sum, np.around(predict_proba[0], decimals=4)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
